```python
ver_immat13 = df_balance_class13.loc['Intangible assets'].round()[0]
ver_immat14 = df_balance_class14.loc['Intangible assets'].round()[0]


sachanlagen13 = df_balance_class13.loc['Tangible assets'].round()[0]
sachanlagen14 = df_balance_class14.loc['Tangible assets'].round()[0]


ver_finanz13 = df_balance_class13.loc['Financial assets'].round()[0]
ver_finanz14 = df_balance_class14.loc['Financial assets'].round()[0]


AV13 = ver_immat13 + sachanlagen13 + ver_finanz13
AV14 = ver_immat14 + sachanlagen14 + ver_finanz14

Ford13 = df_balance_class13.loc['Receivables'].round()[0]
Ford14 = df_balance_class14.loc['Receivables'].round()[0]
```

```python
def Rentabilitaet(Ertrag, Aufwand, EK, FK, Umsatz, Zinsen_FK):
    
    """ Ertrag:  Ertragspositionen des aktuellen Geschäftsjahres
        Aufwand: Aufwandspositionen des aktuellen Geschäftsjahres
        EK: Eigenkaptial des Unternehmens
        FK: Fremdkapital des Unternehmens
        Zinsen_FK: Fremdkapitalszinsen absolut
        
        Kennzahlen:
        Gewinn = Überschuss aus Ertrag - Aufwand vor Steuer
        GK = Gesamtkaptial aus Summe EK + FK
        Rendite(EK, FK, GK, Umsatz): Rendite(n) ist die Verzinsung des eingesetzten Kapitals
        zu einer Bezugsgröße
        Leverage-Effekt: die EK-Rendite steigt mit zunehmenden Fremdkapital(FK), jedoch nur
        solange die GK-Kapitalsverzinsung über der FK-Verzinsung liegt.
        
        """
    Gewinn = Ertrag - Aufwand
    
    rendite_EK = (Gewinn / EK) *100 # Verzisnung des eingesetzten Kapitals
    rendite_FK = (Zinsen_FK / FK) *100 # Ø Verzinsung des Fremdkapitals
    rendite_Umsatz = ( Gewinn / Umsatz)* 100
    rendite_GK = ((Gewinn + Zinsen_FK) / (EK + FK)) * 100
        
    # Leverage Effekt
    if rendite_Umsatz > rendite_FK:
        hebel = 1
    else:
        hebel = 0
    
    return rendite_EK, rendite_FK, rendite_GK, rendite_Umsatz, hebel


def Kapitalstruktur (EK,FK):
    
    ''' EK: Eigenkaptial, FK: Fremdkaptial
        Ermittlung des Anteils EK bzw. FK am Gesamtkapital bzw. Ermittlung 
        des Verschuldungsgrad
    
    '''
    
    GK = EK + FK
    EK_Quote = (EK / GK) * 100
    FK_Quote = (FK / GK) * 100
    Verschuldungsgrad = (FK / EK) * 100
    
    return EK_Quote, FK_Quote, Verschuldungsgrad


def Kennzahlen_Vermögensruktur(AV,EK,FK, Forderungen):
    
    """ Ermittlung der Vermögensstrukur - linke Seite der Bilanz
        AV = Anlagenvermögen
        EK = Eigenkapital
        FK = Fremdkapital
        
        Kennzahlen:
        Anlagenintensität: Anteil des Anlagevermögens am Gesamtvermögen
        Umlauflaufintensivität: kurzfristig im Unternehmen vorhandenes Vermögen
    
        
    """
    
    GV = EK + FK
    UV = GV - AV
    Anlagenintensität = ( AV / GV ) * 100
    Umlaufintensität = ( UV / GV) * 100
    Quote_Ford = (Forderungen / GV ) * 100
    
    return Anlagenintensität, Umlaufintensität, Quote_Ford


def Deckungsgrade(EK, AV, FKlang):
    
    ''''''
    
    Grad_a = ( EK / AV ) * 100
    Grad_b = ((EK + FKlang) / AV ) *100
    
    return Grad_a, Grad_b


def Liquidität(UV, FKkurz, Mittel_liquid):
    
    Grad_1 = ( Mittel_liquid / FKkurz ) * 100
    Grad_2 = ( UV / FKkurz) * 100
    workingcap = ( UV - FKkurz )
    
    return Grad_1, Grad_2, workingcap


    
    
    
```

#### Kennzahlen 13

```python
df_13 = pd.DataFrame(Rentabilitaet(Ertrag13,Aufwand13,EK13,FK13, Umsatz13, Zinsen_FK13), 
         columns=['indicator_13'], index=['Eigenkaptialrentabilität',
                                     'Fremdkaptialrentabilität',
                                     'Return on Investment', 
                                     'Umsatzrentabilität','Leverage-effekt-positiv']).round()


df_13 = df_13.append(pd.DataFrame(Kapitalstruktur(EK13,FK13),columns=['indicator_13'],
            index=['Eigenkaptial Quote', 'Fremdkapital Quote', 'Verschuldungsgrad']))



df_13 = df_13.append(pd.DataFrame(Kennzahlen_Vermögensruktur(AV13,EK13,FK13,Ford13),
                                                  columns=['indicator_13'],
                                                  index=['Anlagen-Intensität', 'Umlauf-Intensität', 'Quote Forderung']))

```
